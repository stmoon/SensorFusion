## 2. Fusion Model
  - $$ x_{k+1} = A_k x_k + B_k v_k $$
  - $$ z^m_k = H^m_k x_k + w^m_k  (m=1,2) $$

```python
import numpy as np

class Measure :
    def __init__(self, Q,R) :
        self.dt = 1
    
        self.A = np.matrix([
                       [1 , self.dt ],
                       [0 ,       1  ]
                    ])
    
        self.B = np.matrix([
                       [self.dt*self.dt/2.0],
                       [self.dt                  ]
                    ])
    
        self.H = np.matrix([
                       [1,0]
                    ])
        self.xk = np.matrix([
                             [0],
                             [0.5]
                    ])
        
        self.Q = Q
        self.R = R
        
    def do(self) :
        vk = np.random.normal(0.0, np.sqrt(self.Q))
        xk_0 = self.A*self.xk + self.B*vk
    
        wk1 = np.random.normal(0.0, np.sqrt(self.R))
        zk1 = self.H*xk_0 + wk1
          
        self.xk = xk_0    
        
        return zk1

## TEST
#zk1, zk2 = measure(0.001, 0.5)
#print zk1[0,0]
```

```python
##########################################
# Test For measurement                                              #
##########################################

import matplotlib.pyplot as plt

data_gt   = []
data_zk1 = []
data_zk2 = []

groundTrue = Measure(1.0*1e-10, 1.0*1e-10)
sensor1 = Measure(0.001, 0.5)
sensor2 = Measure(0.001, 0.5)

for i in range(0,100) :
    
    gt   = groundTrue.do()
    zk1 = sensor1.do()
    zk2 = sensor2.do()
    
    data_gt.append(gt[0,0])
    data_zk1.append(zk1[0,0])
    data_zk2.append(zk2[0,0])
    
plt.plot(data_gt  , label='groundTrue')    
plt.plot(data_zk1, label='zk1')
plt.plot(data_zk2, label='zk2')
plt.legend()
plt.show() 
```

### 2.1. Measurement Fusion Models
- Kalman Filter
    - Model 
        - $$ x_{k+1} = A x_k + w_k $$
        - $$ z_k = H x_k + v_k $$
        -  주의 : $$w_k$$는 상태 변수의 공분산의 행렬값이고, $$ v_k $$는 측정변수의 공분산의 행렬값이다. 
        -  앞에 나온 것과 반대로 되어 있음

```python
import numpy as np
from numpy import linalg

class KalmanFilter() :
    def __init__(self,A,H,Q,R) :              
        self.A = A
        self.H = H
        self.Q = Q
        self.R = R
    
        self.x = np.zeros((A.shape[0],1))
        self.P = np.eye(A.shape[0])
    
    def do(self, z):
        xhat = self.A * self.x;
        Pp = self.A * self.P * self.A.transpose() + self.Q
    
        K  = Pp * self.H.transpose() * linalg.inv(self.H * Pp * self.H.transpose() + self.R)
    
        self.x = xhat + K * ( z - self.H * xhat ) 
        self.P = Pp - K * self.H * Pp         
        
        return self.x, self.P, K
        
```

#### 2.1.1. First Approach

- $$ Z_k = [(z^1_k)^T (z^2_k)^T]^T  
              = \begin{pmatrix}
                    z^1_k & 0 \\
                    0 & z^2_k
                \end{pmatrix} $$
- $$ R_k = \begin{pmatrix}
                    R^1_k & 0 \\
                    0 & R^2_k
                \end{pmatrix} $$

```python
#########################
## TEST For FusionAlgorithm 1 ##
#########################

import numpy as np
import matplotlib.pyplot as plt

data_gt   = []
data_zk1 = []
data_zk2 = []
data_est = []

groundTrue = Measure(1.0*1e-10, 1.0*1e-10)
sensor1     = Measure(0.001, 0.5)
sensor2     = Measure(0.001, 0.5)


dt = 1
A = np.matrix([
                    [1,dt],
                    [0,1]
            ])
H = np.matrix([[1,0],[1,0]])
Q = np.matrix([[1,0],[0,1]])
R = np.matrix([[1,0],[0,1]])
kf          = KalmanFilter(A,H,Q,R)

for i in range(0,100) :
    gt   = groundTrue.do()
    zk1 = sensor1.do()
    zk2 = sensor2.do()

    est,_,_ = kf.do(np.append(zk1,zk2, axis=0) )   
    
    data_gt.append(gt[0,0])
    data_zk1.append(zk1[0,0])
    data_zk2.append(zk2[0,0])    
    data_est.append(est[0,0])
    
    
plt.plot(data_gt  , label='groundTrue')    
plt.plot(data_zk1, label='zk1')
plt.plot(data_zk2, label='zk2')
plt.plot(data_est, label='estimation')
plt.legend()
plt.show()   
```

#### 2.1.2. Second Approach

<img src="files/Images/FusionAlgorithm_1.png" />

  - $$ z_k = z^1_k + R^1_k ( R^1_k + R^2_k ) ^{-1} ( z^2_k - z^1_k ) $$
  

```python
#########################
## TEST For FusionAlgorithm 2 ##
#########################

import numpy as np
import matplotlib.pyplot as plt

data_gt   = []
data_zk1 = []
data_zk2 = []
data_est = []

groundTrue = Measure(1.0*1e-10, 1.0*1e-10)
sensor1     = Measure(0.001, 0.5)
sensor2     = Measure(0.001, 0.5)


dt = 1
A = np.matrix([
                    [1,dt],
                    [0,1]
            ])
H = np.matrix([[1,0]])
Q = np.matrix([[1,0],[0,1]])
R = np.matrix([[1]])
kf          = KalmanFilter(A,H,Q,R)

for i in range(0,100) :
    gt   = groundTrue.do()
    zk1 = sensor1.do()
    zk2 = sensor2.do()

    zk   = zk1 + 0.4*(zk2-zk1)
    est,_,_ = kf.do(zk)   
    
    data_gt.append(gt[0,0])
    data_zk1.append(zk1[0,0])
    data_zk2.append(zk2[0,0])    
    data_est.append(est[0,0])
    
    
plt.plot(data_gt  , label='groundTrue')    
plt.plot(data_zk1, label='zk1')
plt.plot(data_zk2, label='zk2')
plt.plot(data_est, label='estimation')
plt.legend()
plt.show()      
```

### 2.2. The track-to-track fusion models

#### 2.2.1. The track-to-track fusion models

<img src="files/Images/FusionAlgorithm_2.png" />

- $$ \hat{x}_{k|k} = \hat{x}^1_{k|k} + [ P^1_{k|k} - P^{12}_{k|k} ]  [P^1_{k|k} + P^2_{k|k} - P^{12}_{k|k} - P^{21}_{k|k}]^{-1}  \times (\hat{x}^2_{k|k} - \hat{x}^1_{k|k})$$
- $$ P^{12}_{k|k} = ( I - K^1_k H^1_k) A_{k-1} P^{12}_{k-1|k-1} A^T_{k-1} ( I - K^2_k H^2_k )^T +  ( I - K^1_k H^1_k ) B_{k-1} Q_{k-1} B^T_{k-1} (  I - K^2_k H^2_k )^T $$


```python
#############################
## TEST For FusionAlgorithm 2.2.1   ##
## The track-to-track fusion models  ##
#############################

import numpy as np
import matplotlib.pyplot as plt

data_gt   = []
data_zk1 = []
data_zk2 = []
data_est = []

groundTrue = Measure(1.0*1e-10, 1.0*1e-10)
sensor1     = Measure(0.001, 0.5)
sensor2     = Measure(0.001, 0.5)


dt = 1
A = np.matrix([
                    [1,dt],
                    [0,1]
            ])
B = np.matrix([
               [dt*dt/2.0],
               [dt                  ]
            ])

H = np.matrix([[1,0]])
Q = np.matrix([[1,0],[0,1]])
R = np.matrix([[1]])
kf1 = KalmanFilter(A,H,Q,R)
kf2 = KalmanFilter(A,H,Q,R)

P12_ = np.eye(2)
    
for i in range(0,100) :
    gt   = groundTrue.do()
    
    zk1 = sensor1.do()
    xhat1,P1,K1 = kf1.do(zk1)
    
    zk2 = sensor2.do()
    xhat2,P2,K2 = kf2.do(zk2)
    
    I = np.eye(2)

    P12 =   (I - K1*H)*A*P12_*A.transpose() * (I - K2*H).transpose() + (I - K1*H)*B*R*B.transpose() * (I - K2*H).transpose()
    P12_ = P12
    
    xhat = xhat1 + (P1 - P12) * linalg.inv(P1 + P2 - P12 - P12.transpose()) * (xhat2 - xhat1)
    
    data_gt.append(gt[0,0])
    data_zk1.append(zk1[0,0])
    data_zk2.append(zk2[0,0])    
    data_est.append(xhat[0,0])
    
    
plt.plot(data_gt  , label='groundTrue')    
plt.plot(data_zk1, label='zk1')
plt.plot(data_zk2, label='zk2')
plt.plot(data_est, label='estimation')
plt.legend()
plt.show()      
```

#### 2.2.2. The modified track-to-track fusion algorithm (MTF)

<img src="files/Images/FusionAlgorithm_3.png" />

- $$ \bar{z}^m_{k|k-1} = H^m_k \hat{x}_{k|k-1} $$
- $$ \hat{x}_{k|k} = \hat{x}_{k|k-1} + P^m_{k|k-1}(x,z) P^{m}_{k|k-1} (z,z)^{-1} [ z^m_k - \bar{z}^m_{k|k-1} ]  $$



```python
########################################
## TEST For FusionAlgorithm 2.2.2                         ##
## The modified track-to-track fusion models (MTF) ##
########################################


import numpy as np
from numpy import linalg

class FusionKalmanFilter() :
    def __init__(self,A,H,Q,R) :              
        self.A = A
        self.H = H
        self.Q = Q
        self.R = R
    
        self.x = np.zeros((A.shape[0],1))
#         self.P = np.eye(A.shape[0])
        self.P = np.matrix([[0.01, 0],[0,0.01]])
    
    def calculate(self, z1, z2):       
        
        lamda = np.matrix([
                       [1/2.0],
                       [1       ]
                    ])
        
        # 1. 추정값과 오차 공분산 예츨
        xhat = self.A * self.x;        
        Ppri = self.A*self.P*self.A.T + lamda*self.Q*lamda.T
        
        # 2. 칼만이득 계산
        K1  = Ppri * self.H.T * (self.H * Ppri * self.H.T + self.R).T
        K2  = Ppri * self.H.T * (self.H * Ppri * self.H.T + self.R).T        

        # 3. 추정값 계산
        x1 = xhat + K1 * ( z1 - self.H * xhat )         
        x2 = xhat + K2 * ( z2 - self.H * xhat )         

        # 4. 오차공분산 계산
        I         = np.matrix(np.eye(2))
        P1      = (I - K1*self.H)*Ppri
        P2      = (I - K2*self.H)*Ppri
        P12    = (I - K1*self.H)*Ppri*(np.eye(2) - K2*self.H).T
        P21    = P12.T
        Ppart  = P1 + P2 - P12 - P21      
        if np.linalg.det(Ppart) == 0 :       
            Ppart = I
        
            
        self.P  = P1 - (P1 - P12)*Ppart.I*(P1-P21)
                                                           
        # 5. Fusion bewteen x1 and x2
        self.x = x1 + (P1-P12)*Ppart*(x2-x1)           
            
        return self.x
    


###################################################
import numpy as np
import matplotlib.pyplot as plt

data_gt   = []
data_zk1 = []
data_zk2 = []
data_est = []

groundTrue = Measure(1.0*1e-10, 1.0*1e-10)
sensor1     = Measure(0.1, 0.5)
sensor2     = Measure(0.001, 0.5)


dt = 1
A = np.matrix([
                    [1,dt],
                    [0,1]
            ])
B = np.matrix([
               [0],
               [0]
            ])

H = np.matrix([[1,0]])
Q = 0.001
R = np.matrix([[0.5]])

fkf = FusionKalmanFilter(A,H,Q,R)
    
for i in range(0,100) :
    gt   = groundTrue.do()
    
    zk1 = sensor1.do()  
    zk2 = sensor2.do()
    xhat= fkf.calculate(zk1,zk2)
    
    data_gt.append(gt[0,0])
    data_zk1.append(zk1[0,0])
    data_zk2.append(zk2[0,0])    
    data_est.append(xhat[0,0])
    
    
plt.plot(data_gt  , label='groundTrue')    
plt.plot(data_zk1, label='zk1')
plt.plot(data_zk2, label='zk2')
plt.plot(data_est, label='estimation')
plt.legend()
plt.show()      
```
